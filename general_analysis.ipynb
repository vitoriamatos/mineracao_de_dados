#### Importando as dependencias

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Incluindo a Base de Dados

```python
database = "data/database/database_fit.csv"
df = pd.read_csv(database, delimiter=',', encoding = "ISO-8859-1")
```

#### Análise exploratória Inicial

```python
girls = df.query("alunosexo == 'F' and  nmodennome == 'EDUCACAO  JOVENS E ADULTOS' and (situ == 'D' or situ == 'RF' or situ == 'NC')")
girls_total = df.query("alunosexo == 'F'")
percentual_girls = round((len(girls)/len(girls_total)),3)

boys = df.query("alunosexo == 'M' and nmodennome == 'EDUCACAO  JOVENS E ADULTOS' and (situ == 'D' or situ == 'RF' or situ == 'NC')")
boys_total = df.query("alunosexo == 'M'")
percentual_boys = round((len(boys)/len(boys_total)),3)
```

```python
print("================INFO MENINAS================")

print('TOTAL ALUNAS: '+ str(len(girls_total)))
print('TOTAL EVASÃO FEMININA: '+ str(len(girls)))
print('PERCENTUAL EVASÃO: '+ str(percentual_girls))

print("================INFO MENINOS================")

print('TOTAL ALUNOS: '+ str(len(boys_total)))
print('TOTAL EVASÃO MASCULINA: '+ str(len(boys)))
print('PERCENTUAL EVASÃO: '+ str(percentual_boys))
```

```python
student_evasion = df.query("nmodennome == 'EDUCACAO  JOVENS E ADULTOS' and (situ == 'D' or situ == 'RF' or situ == 'NC')")

student_evasion['alunosexo'].value_counts().plot(kind='bar')
plt.title("Comparação de evasão entre meninos e meninas")
plt.show()
```

```python
student_evasion = df.query("alunosexo == 'F'")

student_evasion['situ'].value_counts().plot(kind='bar')
plt.title("Comparação de evasão entre meninas")
plt.show()
```

Análise de variaveis

```python
database = "data/database/data.csv"
data = pd.read_csv(database, delimiter=',', encoding = "ISO-8859-1")
```

```python
girls = data.query("genero == 1 and  serie == 2 and resultado == 1")
girls['nivelbairro'].value_counts().plot(kind='bar')
plt.title("Distribuição de evasão por niveis de bairro")
plt.show()
```

Calcular o percentual de evasão por cada bairro

```python
bairro_classe_alta_com_evasao = data.query("genero == 1 and  serie == 2 and resultado == 1 and nivelbairro == 1")
bairro_classe_alta_sem_evasao = data.query("genero == 1 and  serie == 2 and resultado == 0 and nivelbairro == 1")

percentual_evasao_classe_alta = round(len((bairro_classe_alta_com_evasao)/ len(bairro_classe_alta_sem_evasao)),3)

print('PERCENTUAL EVASÃO CLASSE ALTA: '+ str(percentual_evasao_classe_alta) + '%')


bairro_classe_media_com_evasao = data.query("genero == 1 and  serie == 2 and resultado == 1 and nivelbairro == 2")
bairro_classe_media_sem_evasao = data.query("genero == 1 and  serie == 2 and resultado == 0 and nivelbairro == 2")

percentual_evasao_classe_media = round(len((bairro_classe_media_com_evasao)/ len(bairro_classe_media_sem_evasao)),3)

print('PERCENTUAL EVASÃO CLASSE MÉDIA: '+ str(percentual_evasao_classe_media) + '%')

bairro_classe_baixa_com_evasao = data.query("genero == 1 and  serie == 2 and resultado == 1 and nivelbairro == 3")
bairro_classe_baixa_sem_evasao = data.query("genero == 1 and  serie == 2 and resultado == 0 and nivelbairro == 3")

percentual_evasao_classe_baixa = round(len((bairro_classe_baixa_com_evasao)/ len(bairro_classe_baixa_sem_evasao)),3)

print('PERCENTUAL EVASÃO CLASSE BAIXA: '+ str(percentual_evasao_classe_baixa) + '%')

bairro_classe_periferia_com_evasao = data.query("genero == 1 and  serie == 2 and resultado == 1 and nivelbairro == 4")
bairro_classe_periferia_sem_evasao = data.query("genero == 1 and  serie == 2 and resultado == 0 and nivelbairro == 4")

percentual_evasao_classe_periferia = round(len((bairro_classe_periferia_com_evasao)/ len(bairro_classe_periferia_sem_evasao)),3)

print('PERCENTUAL EVASÃO PERIFERIA: '+ str(percentual_evasao_classe_periferia) + '%')
```

#### Conclusão

Pode-se concluir de maneira geral que, a taxa de evasão entre `mulheres no ensino para jovens e adultos` é superior do que para os homens, e ocorrre com mais frequencia em bairros de periferia e classe baixa. Apesar disso, a **taxa de evasão feminina** corresponde a apenas 0.011%, indicando que felizmente, apenas um pequeno número de mulheres evadem da escola.
